```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/time series data.csv')
```

```python
data.head(5)
```

```python
data.shape
```

```python
# Seaborn style
sns.set_style('whitegrid')


plt.figure(figsize=(10, 6))
sns.lineplot(x='t', y='ProductP1', data=data, marker='o', color='red',label='ProductP1')

plt.title('ProductP1 Sold Over Time (t)')
plt.xlabel('Date')
plt.ylabel('Product Sold')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='t', y='ProductP1', data=data, marker='*', color='blue', label='ProductP1')
sns.lineplot(x='t', y='ProductP3', data=data, marker='o', color='orange', label='ProductP3')

plt.title('ProductP1 & ProductP3 Over Sales Time (t)')
plt.xlabel('Time')
plt.ylabel('ProductSold')
plt.legend()

plt.show()
```

```python
sns.set_style('whitegrid')

product_to_plot = ['ProductP1', 'ProductP2', 'ProductP3', 'ProductP4', 'ProductP5']

plt.figure(figsize=(10,8))
for products in product_to_plot:
    sns.lineplot(x='t',y=products, data = data,marker='*',label=products)

plt.title('Scatter Plot for Products Over Time')
plt.xlabel('Time')
plt.ylabel('Product Sold')
plt.legend()

plt.show()

```

```python
sns.set_style('whitegrid')

product_to_plot = ['ProductP1', 'ProductP2', 'ProductP3', 'ProductP4', 'ProductP5']

plt.figure(figsize=(10,8))
for products in product_to_plot:
    sns.scatterplot(x='t',y=products, data = data,marker='o',label=products)

plt.title('Scatter Plot for Products Over Time')
plt.xlabel('Time')
plt.ylabel('Product Sold')
plt.legend()

plt.show()

```

```python
# Seaborn style
sns.set_style('whitegrid')


plt.figure(figsize=(10, 6))
sns.lineplot(x='t', y='TotalSales', data=data, marker='o', color='black',label='Total Sale')

plt.title('ProductP1 Sold Over Time (t)')
plt.xlabel('Date')
plt.ylabel('Product Sold')
plt.legend()

plt.show()
```
