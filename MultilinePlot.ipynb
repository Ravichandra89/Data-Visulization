```python
# Importing libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Importing the dataset file using pandas
data = pd.read_csv('/content/Cost_of_Living_Index_2022.csv')
```

```python
# Printing the first 5 line of the data
data.head()
```

```python
data.shape
```

```python
# Selcting the attributes
attributes = ['Rank','Cost of Living Index','Rent Index']
```

```python
# Defining different colors, markers and Legend
colors = ['b', 'g', 'r']
markers = ['o', '*', '^']

legend_labels = ['Period','Revenue','Quantity','Average_cost']
```

```python
# Now creating multiline plot using plt.plot()
plt.figure(figsize=(10, 6))

plt.title('Multiline Plot on Cost of Living index data')

for i, attribute in enumerate(attributes):
    plt.plot(data[attribute], color=colors[i], marker=markers[i], label=legend_labels[i])
```
