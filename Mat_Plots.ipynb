```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Generating DataFrame
data = {
    'A': np.random.normal(0, 1, 50),
    'B': np.random.normal(2, 1, 50),
    'C': np.random.normal(1, 1, 50),
}

df = pd.DataFrame(data)
```

```python
# Printing the Frame
print(df)
```

```python
# Making plots Between Attribute B and C
df_select = df[['B','C']]

fig, axes = plt.subplots(nrows=2,ncols=4,figsize=(20,9))


# Plots
df_select.plot(kind='line',ax=axes[0,0])
axes[0,1].set_title("Box Plot Between B & C")

# Box plot
sns.boxplot(data=df_select,ax=axes[0,1])
axes[0,1].set_title("Box plot Between B and C")

# Histogram plot between B And C
df_select.plot(kind='hist', alpha=0.5, bins=20, stacked=True, ax=axes[0, 2])
axes[0, 2].set_title('Histogram Between B and C')

# Violin plot between B and C
sns.violinplot(data=df_select, ax=axes[1, 0])
axes[1, 0].set_title('Violin Plot Between B and C')

# Scatter plot
axes[1, 1].scatter(df_select['B'], df_select['C'])
axes[1, 1].set_title('Scatter Plot Between B and C')
axes[1, 1].set_xlabel('B')
axes[1, 1].set_ylabel('C')

# Heatmap
sns.heatmap(df_select.corr(), annot=True, cmap="copper",ax=axes[1,2])
axes[1,2].set_title('Heatmap Between B and C')


plt.tight_layout()
plt.show()
```
