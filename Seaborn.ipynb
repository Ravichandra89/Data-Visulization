# ***SEABORN PLOTS***

```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = x*2;

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Chart")

plt.plot(x,y,marker="o",linestyle="dotted")
plt.show()
```

```python
# Multi graph with Same axis

x = np.array([1,2,3,4,5])
y = x*2

plt.title("Multi Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.plot(x,y)
plt.show()

plt.figure()
d1 = np.array([2,4,6,8])
d2 = np.array([3,5,7,9])

# Labeling
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("New Graph with d1 & d2")

plt.plot(d1,d2,ls="dashed",color="r")
plt.show()

# Using figure() fun
```

```python
# Multiline plots
x = np.array([1,2,3,4,5])
y = x*2;

plt.plot(x,y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

# second plot with x1 and y1 data
plt.plot(x1, y1,color='r')

plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')
plt.show()
```
